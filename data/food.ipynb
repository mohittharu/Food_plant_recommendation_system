```python
# columns_with_outliers = ['pH', 'Potassium']


# def remove_outliers(data, columns):
#     for column in columns:

#         Q1= data[column].quantile(0.25)
#         Q3=data[column].quantile(0.75)

#         IQR = Q3 - Q1
#         outlier_step = 1.5 * IQR
        
#         remove_indices = data[(data[column])]

import pandas as pd

columns_with_outliers = ['pH','Potassium','Soil EC', 'T.S.P']

def remove_outliers(data, columns):
    for column in columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
  
        IQR = Q3 - Q1
      
        outlier_step = 1.5 * IQR
        
        outliers_indices = data[(data[column] < Q1 - outlier_step) | (data[column] > Q3 + outlier_step)].index
    
        data.drop(outliers_indices, inplace=True)
    return data


data_no_outliers = remove_outliers(data, columns_with_outliers)


# print("DataFrame after removing outliers:")
# print(dset1_no_outliers)
```

```python
target_count = 3300
undersampled_data = []

for type in data['Plant Type'].unique():
    plant_type_data =data[data["Plant Type"]== type]

    if len(plant_type_data)> target_count:

        undersampled_data.append(plant_type_data.sample(n= target_count, random_state=42))

    else:

        undersampled_data.append(plant_type_data)

undersampled_data= pd.concat(undersampled_data)


undersampled_data.reset_index(drop=True, inplace=True)

print("Undersampled DataFrame")
print(undersampled_data)


undersampled_data.to_csv("undersampled_data.csv", index= False)

```

```python
data = pd.read_csv('undersampled_data.csv')
data
```

```python
# Normalization 

from sklearn.preprocessing import StandardScaler

columns_to_normalize = ['pH', 'Phosphorus', 'Potassium','Soil EC', 'T.S.P']

scaler = StandardScaler()

data[columns_to_normalize] = scaler.fit_transform(data[columns_to_normalize])
```

```python
data.head(5)
```

```python
# Data Modeling 

from sklearn.model_selection import train_test_split

features = data[['pH', 'Phosphorus', 'Potassium','Soil EC', 'T.S.P']]
target = data['Plant Type']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(features,target, test_size=0.2, random_state=42)

```

```python
# Decision Trees 

from sklearn.ensemble import  RandomForestClassifier
from sklearn import metrics

RandomForest = RandomForestClassifier(n_estimators =70 , random_state= 42)

RandomForest.fit(xtrain, ytrain)


ypred = RandomForest.predict(xtest)


accuracy = metrics.accuracy_score(ypred, ytest )
```

```python
print(accuracy)
```
